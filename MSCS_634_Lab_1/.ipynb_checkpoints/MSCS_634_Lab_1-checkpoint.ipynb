```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Real_Estate_Sales.csv')
```

```python
x_col = "SalePrice"
y_col = "TotalFinishedArea"

filtered_df = df[[x_col, y_col]].dropna()

plt.figure(figsize=(8, 6))
plt.scatter(filtered_df[x_col], filtered_df[y_col], alpha=0.5)
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.title(f"{y_col} vs {x_col}")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
grouped = df.groupby("AssrLandUse")["SalePrice"].mean().dropna().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
grouped.plot(kind='bar')
plt.ylabel("Average SalePrice")
plt.xlabel("Assessed Land Use")
plt.title("Average Sale Price by Land Use Category")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df2 = pd.read_csv('Real_Estate_Sales_Cleaned.csv')
df2.info()
df2.describe()
```

```python
min_values = df2.min(numeric_only=True)
max_values = df2.max(numeric_only=True)
mean_values = df2.mean(numeric_only=True)
median_values = df2.median(numeric_only=True)
mode_values = df2.mode(numeric_only=True).iloc[0]  # .mode() returns a DataFrame

# Display the results together
summary_stats = pd.DataFrame({
    'Min': min_values,
    'Max': max_values,
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values
})

summary_stats
```

```python
numeric_df = df2.select_dtypes(include=['float64', 'int64'])

min_vals = numeric_df.min()
max_vals = numeric_df.max()
range_vals = max_vals - min_vals
q1 = numeric_df.quantile(0.25)
q2 = numeric_df.quantile(0.50)
q3 = numeric_df.quantile(0.75)
iqr = q3 - q1
variance = numeric_df.var()
std_dev = numeric_df.std()

stats_summary = pd.DataFrame({
    "Min": min_vals,
    "Max": max_vals,
    "Range": range_vals,
    "Q1 (25%)": q1,
    "Median (Q2)": q2,
    "Q3 (75%)": q3,
    "IQR": iqr,
    "Variance": variance,
    "Std Dev": std_dev
})

stats_summary
```
