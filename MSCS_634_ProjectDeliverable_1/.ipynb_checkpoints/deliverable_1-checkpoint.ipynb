```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataset/general_dentistry_approintments_original.csv')
df.info()
df.describe()
```

```python
# Data Distribution (Histograms)
df2 = pd.read_csv('dataset/general_dentistry_approintments_cleaned.csv')

df2.hist(figsize=(12, 10), bins=20, edgecolor='black')
plt.suptitle("Distribution of Numeric Variables")
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection (Boxplots)
numeric_cols = df2.select_dtypes(include='number').columns

for col in numeric_cols:
    plt.figure(figsize=(6, 2))
    plt.boxplot(df2[col].dropna(), vert=False)
    plt.title(f'Outliers in {col}')
    plt.xlabel(col)
    plt.grid(True)
    plt.show()
```

```python
# Feature Relationships
plt.scatter(df2['patient_height'], df2['patient_weight'], alpha=0.7)
plt.title("Patient Height vs. Weight")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(True)
plt.show()
```
