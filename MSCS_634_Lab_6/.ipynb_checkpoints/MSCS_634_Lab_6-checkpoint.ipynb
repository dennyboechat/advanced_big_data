```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
df = pd.read_csv('Online Retail.csv', encoding='ISO-8859-1')

# Step 3: Rename corrupted column name
df.rename(columns={'ï»¿InvoiceNo': 'InvoiceNo'}, inplace=True)

# Step 4: Convert InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m/%d/%y %H:%M', errors='coerce')

# Step 5: Drop rows with missing values in key fields
df = df.dropna(subset=['CustomerID', 'Description'])

# Step 6: Remove rows with Quantity <= 0 or UnitPrice <= 0
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Step 7: Drop duplicate rows
df = df.drop_duplicates()

# Step 8: Reset index
df.reset_index(drop=True, inplace=True)

# --- Barplot: Top 10 most frequent items ---
top_items = df['Description'].value_counts().nlargest(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_items.values, y=top_items.index)
plt.title('Top 10 Most Frequently Occurring Items')
plt.xlabel('Frequency')
plt.ylabel('Item Description')
plt.tight_layout()
plt.show()

# --- Heatmap: Item co-occurrence (Top 20 Items) ---
# Filter dataset to top 20 items only
top_20_items = df['Description'].value_counts().nlargest(20).index
filtered_df = df[df['Description'].isin(top_20_items)]

# Create a binary transaction matrix
basket = (
    filtered_df
    .groupby(['InvoiceNo', 'Description'])['Quantity']
    .sum()
    .unstack()
    .fillna(0)
)
basket[basket > 0] = 1  # Convert to binary

# Compute co-occurrence matrix
co_occurrence = basket.T.dot(basket)

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(co_occurrence, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Item Co-occurrence Heatmap (Top 20 Items)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

Mining Using Apriori 

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Convert to boolean to avoid warning
basket = basket.astype(bool)

# Generate frequent itemsets
frequent_itemsets = apriori(basket, min_support=0.01, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)

# Sort by lift
rules_sorted = rules.sort_values(by='lift', ascending=False)

# Show top rules
rules_sorted.head()
```

Set appropriate support thresholds based on dataset size and characteristics

```python
# Create basket (transaction x item matrix for top items or by country if needed)
top_items = df['Description'].value_counts().nlargest(30).index
filtered_df = df[df['Description'].isin(top_items)]

basket = (
    filtered_df
    .groupby(['InvoiceNo', 'Description'])['Quantity']
    .sum()
    .unstack()
    .fillna(0)
)

# Convert to binary and boolean (required by Apriori)
basket[basket > 0] = 1
basket = basket.astype(bool)

# Define thresholds to test
support_levels = [0.05, 0.02, 0.01, 0.005, 0.001]
results = []

for support in support_levels:
    # Generate frequent itemsets
    itemsets = apriori(basket, min_support=support, use_colnames=True)
    # Generate association rules
    rules = association_rules(itemsets, metric='lift', min_threshold=1.0)
    # Save results
    results.append({
        'Support': support,
        'Frequent Itemsets': len(itemsets),
        'Association Rules': len(rules)
    })

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Plot results
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df.melt(id_vars='Support'), x='Support', y='value', hue='variable')
plt.title('Effect of Support Threshold on Itemsets and Rules')
plt.ylabel('Count')
plt.xlabel('Minimum Support')
plt.yscale('log')  # optional: use log scale for better visibility
plt.tight_layout()
plt.show()

# Show detailed result table
results_df
```

FP-Growth and Apriori comparison

# Efficiency (Speed & Memory)
## Apriori:

Generates candidate itemsets and scans the dataset multiple times.

Becomes slow with large datasets or low support thresholds.

Time complexity grows exponentially with the number of unique items.

## FP-Growth:

Builds a compact FP-Tree representation of the dataset.

Does not generate candidates explicitly.

Usually much faster and more memory-efficient for large datasets.

# Output (Itemsets & Rules)
Both algorithms produce the same frequent itemsets and association rules if the same min_support and min_confidence thresholds are applied.

FP-Growth may handle lower min_support values that Apriori would take too long to process.

```python
import time
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules

# Convert basket to boolean
basket = basket.astype(bool)

# --- Apriori ---
start_time = time.time()
frequent_itemsets_apriori = apriori(basket, min_support=0.01, use_colnames=True)
apriori_time = time.time() - start_time

# --- FP-Growth ---
start_time = time.time()
frequent_itemsets_fpgrowth = fpgrowth(basket, min_support=0.01, use_colnames=True)
fpgrowth_time = time.time() - start_time

# --- Output comparison ---
print(f"Apriori: {len(frequent_itemsets_apriori)} itemsets, Time: {apriori_time:.2f} seconds")
print(f"FP-Growth: {len(frequent_itemsets_fpgrowth)} itemsets, Time: {fpgrowth_time:.2f} seconds")

# Preview outputs
print("\nTop Apriori itemsets:\n", frequent_itemsets_apriori.head())
print("\nTop FP-Growth itemsets:\n", frequent_itemsets_fpgrowth.head())
```

Rules interpretation and insights about meaningful patterns

The association rules generated using the FP-Growth algorithm help uncover patterns in customer purchasing behavior by analyzing metrics like support, confidence, and lift. Support tells us how often a combination of items appears in all transactions, while confidence shows how likely customers are to buy one item given they’ve already bought another. Lift is especially useful because it reveals how much more likely two items are to be bought together compared to random chance—a lift greater than 1 indicates a positive relationship. For example, a rule with high lift and confidence might show that customers who buy a set of cake tins are significantly more likely to also purchase matching tea canisters. These kinds of insights are useful for making decisions in retail, such as bundling products, setting up cross-selling strategies, or targeting specific customer segments. By focusing on rules with strong metrics, we can find meaningful patterns that can improve marketing efforts and enhance the shopping experience.

Key insights from the generated rules

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules

# Ensure basket is boolean
basket = basket.astype(bool)

# Run FP-Growth and generate rules
fp_itemsets = fpgrowth(basket, min_support=0.01, use_colnames=True)
rules_fp = association_rules(fp_itemsets, metric='confidence', min_threshold=0.5)

# Plot: Confidence vs. Lift
plt.figure(figsize=(10, 6))
sns.scatterplot(data=rules_fp, x='confidence', y='lift', size='support',
                hue='support', palette='viridis', sizes=(20, 200))
plt.title('Confidence vs Lift for Association Rules')
plt.xlabel('Confidence')
plt.ylabel('Lift')
plt.legend(title='Support', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Apriori and FP-Growth comparison

When comparing Apriori and FP-Growth on our dataset, FP-Growth turned out to be noticeably faster and more efficient. Both algorithms produced the same frequent itemsets and association rules since they follow the same logic for identifying patterns, but the difference lies in how they process the data. Apriori had to scan the dataset multiple times and generate a large number of candidate itemsets, which made it slower and more memory-intensive. On the other hand, FP-Growth was much quicker because it builds a compact structure called an FP-tree and avoids the costly candidate generation step altogether. This makes FP-Growth more scalable and better suited for larger datasets or situations where we want to explore less frequent patterns using lower support thresholds. Overall, FP-Growth was the better choice for our dataset due to its speed and efficiency.
